```python
import numpy as np
import matplotlib.pyplot as pp
pp.style.use('fivethirtyeight')
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10, 40)
x
```

```python
sinx = np.sin(x)
```

```python
sinx
```

```python
pp.plot(x, sinx)
```

```python
cosx = np.cos(x)
```

```python
pp.plot(x, sinx)
pp.plot(x, cosx)
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2
pp.plot(x, y)
pp.plot(x, z)
pp.legend(['sinx * cosx', 'cosx**2 - sinx**2'])
```

```python
np.dot(sinx, cosx) #dot product give a scalar value : x.y = x1y1 + x2y2 + x3y3 + .... (+ve value acute angle; -ve obtuse)
```

```python
mat_outer = np.outer(sinx, cosx) #the outer product of u and v is w=uvT.
mat_outer
```
